```python
# Achromatic Delay Scaling — Dark Energy Principle

import numpy as np
import matplotlib.pyplot as plt

# Parameters (normalized)
c_DE = 1.0          # invariant speed baseline
mu = 0.1            # mean offset ⟨n_γ - 1⟩
L = np.linspace(0, 10, 200)  # path length (arbitrary units)

# Delay: Δt = (L / c_DE) * μ
delta_t = (L / c_DE) * mu

# Plot
plt.figure(figsize=(7,4.5))
plt.plot(L, delta_t, color="#1f77b4", linewidth=2)
plt.xlabel("Path length L (arb. units)")
plt.ylabel("Achromatic delay Δt (arb. units)")
plt.title("Achromatic delay scales linearly with path length")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
